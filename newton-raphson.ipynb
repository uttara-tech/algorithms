```python
import math
import numpy as np
from sympy import Derivative, Symbol, factor
import matplotlib.pyplot as plot
```

```python
def f(x):
    return -x**3+6*x**2+15*x+4

x=Symbol('x')

f(x)
```

```python
#finding derivative

Derivative(f(x),x)

df=Derivative(f(x),x).doit()
df
```

```python
#factorising the derivative to find critical points

Derivative(f(x),x)

df=Derivative(f(x),x).doit()
df

factor(df)
```

```python
# Evaluating f(x) at x=5

f(x).evalf(subs={x:5})
```

```python
# Evaluating f(x) at x=-1

f(x).evalf(subs={x:-1})
```

```python
#Plot the function

x=np.linspace(-120,120)
y=f(x)

figCriticalPoints=plot.figure()
ax=plot.gca()
_=ax.set_title("y=f(x)")
_=ax.plot(x,y)
plot.show()
```

```python
# Newton-Rapshon method - it is an iterative method for finding the root of a differrentiable function f(x), 
# which are solutions to the equation f(x)=0. It is represented mathematically as:
# x1 = x0 - f(x0)/f'(x0)
# In this example, we use a built-in function from scipy to find root directly


from scipy import optimize

x=np.linspace(start=-4, stop=6)
ax=plot.gca()
ax.grid(True)
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')

_=ax.set_xlim(-5,7)
_=ax.set_ylim(-200,200)

_=ax.set_title("")
_=ax.plot(x,f(x))

# built-in optimization function
# This line of code represents the defined function and (modifiable) initial root of this function. 
# From this point on , the method iteratively enhances the approximation until it approaches a root.
root=optimize.newton(func=f, x0=2)

print('Root:', root)

plot.show()
```

```python
# Plotting trignometric functions

theta=np.linspace(start=0, stop=6*np.pi, num=100)
sine=np.sin(theta)
cosine=np.cos(theta)
_=plot.plot(theta, sine, theta, cosine)
plot.show()
```
